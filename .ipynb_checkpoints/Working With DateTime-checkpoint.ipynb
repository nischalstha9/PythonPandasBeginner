##### We can also convert to date time during CSV import

```python
from datetime import datetime
df_dateV=pd.read_csv('data/DateTime.csv', parse_dates=['Date'], date_parser=lambda x: datetime.strptime(x,'%Y-%m-%d %I-%p'))
```
