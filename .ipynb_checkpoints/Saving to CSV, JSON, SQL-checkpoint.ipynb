```python
import pandas as pd #importing pandaas
df = pd.read_csv('data/survey_results_public.csv') # importing csv as data frame
schema_df = pd.read_csv('data/survey_results_schema.csv') # importing schema_csv as data frame_schema
pd.set_option('display.max_columns',85)#option to view all columns in dataframe
pd.set_option('display.max_rows',85)#option to view all rows in dataframe
df.set_index('Respondent', inplace=True)
schema_df.set_index(('Column'), inplace=True) #inplace makes indexing permanent
```

```python
filtr = (df['Country']=='Nepal')&(df['LanguageWorkedWith'].str.contains('Python'))
```

```python
Nepal_df = df[filtr]
```

```python
Nepal_df
```

```python
Nepal_df.to_csv('data/Nepal_StackOverFlow_knows_Python.csv')
```

```python
Nepal_df.to_csv('data/Nepal_StackOverFlow_knows_Python.tsv', sep='\t')#tsv is Tab Seperated Values, sep='\t' is required for reading TSV files
```

```python
Nepal_df.to_json('data/Nepal_StackOverFlow_knows_Python.json')
```
