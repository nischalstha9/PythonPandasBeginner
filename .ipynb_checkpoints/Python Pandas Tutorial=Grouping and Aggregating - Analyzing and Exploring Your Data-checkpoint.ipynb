### SImple Plotting of above data

```python
import matplotlib.pyplot as plt
import numpy as np
keys = df['SocialMedia'].value_counts().keys()
values = df['SocialMedia'].value_counts().tolist()
mean = np.mean(values)
fig, ax = plt.subplots()
ax.barh(keys, values)
ax.set_xlabel('Number of Surveyers')
ax.set_ylabel('Social Media Platform')
ax.set_title('Which social media do you use?')
```

#### Here we  found how many percent knows Python using value_counts method now lets do sth more better

```python
country_respondants = df['Country'].value_counts()
```

```python
country_python_user = country_grp['LanguageWorkedWith'].apply(lambda x:x.str.contains('Python').sum())
```

```python
#now we will concatenate country_respondants and country_python_user
```

```python
countryPyUserDF = pd.concat([country_respondants,country_python_user], axis='columns')
```

```python
countryPyUserDF
```

```python
countryPyUserDF.rename(columns = {'LanguageWorkedWith':'DevKnowsPython','Country':'totalDev'}, inplace=True)
```

```python
countryPyUserDF['perKnowPy']=(countryPyUserDF['DevKnowsPython']/countryPyUserDF['totalDev'])*100
```

```python
countryPyUserDF.sort_values(by=['perKnowPy'], ascending = False)
```
