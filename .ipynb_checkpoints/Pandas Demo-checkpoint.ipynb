# Making Custom Filter 

```python
countries = ['United States','Germany','Finland','India','China']
```

```python
filtr_by_country = df['Country'].isin(countries)#setting custom filter
```

```python
df.loc[filtr_by_country&filtr_high_salary,['Country','ConvertedComp']]#merged two filters
```

```python
filtr_by_language = df['LanguageWorkedWith'].str.contains('Python', na=False)#filtering if data contains
```

```python
df.loc[filtr_by_language, 'LanguageWorkedWith']
```

```python
filtr_by_language = df['LanguageWorkedWith'].str.contains(r'^(?=.*Ruby)(?=.*Python)(?=.*Clojur)(?=.*TypeScript)(?=.*Java)(?=.*Assembly)(?=.*PHP)',regex=True, na=False)#filtering if data contains multiple AND Cased(MUST CONTAIN EVERY REGEX STRING) STR 
```

```python
filtr_by_language = df['LanguageWorkedWith'].str.contains('Python|Ruby', na=False)#filtering if data contains any of the string in dataFrame
```

```python
df[filtr_by_language]
```
