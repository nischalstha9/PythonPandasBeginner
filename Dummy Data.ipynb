```python
import pandas as pd
import numpy as np
people = {
    'first':['corey','john','Jane','Chris',np.nan,None,'NA'],
    'last':['Boustany','Smith','Evans','Hemsworth',np.nan,np.nan,'Missing'],
    'email':['CoreyBoustany@yahoo.com','JohnSmith@E.com','Janeevan@y.com',None,np.nan,'anonymous@yaho.com','NA'],
    'age':['33','55','63','36',None,None,'Missing']
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis='index', how='all')#ax-s=columns for dropping col
```

```python
df.dropna(axis='index', how='all', subset = ['email'])#checks email field only
```

```python
df.dropna(axis='index', how='all', subset = ['last','email'])#AND logic on subset
```

```python
df.replace('NA',np.nan, inplace=True)
df.replace('Missing',np.nan, inplace=True)#Converting Custom Missing values to np.Nan
```

```python
df
```

```python
df.dropna(axis='index', how='all',subset=['last','email'])
```

```python
df.isna()
```

```python
df.fillna('0')#fillin NaN
```
